- We need to check the medals column too!
- First we need to remove the **NaN** values and then in those records we need to filter 2016 records

```python
df.Medal.unique()
```

- We need to separate 2016 data from the rest 

```python
df.Year.unique()
```

- Confirmed that there are records of year 2016.
- We need to pull those records and put it into a separate dataframe

- We need to drop those rows where value in medal column in nan

- .dropna() drops the rows with null values in a specific column.

```python
mdf = df.dropna(subset=['Medal'])
```

```python
mdf.head()
```

```python
mdf.Medal.unique()
```

- We managed to drop null values from Medal and kept those records that have won medals

- We need to retrieve records that are of year 2016.

```python
mdf16 = mdf[mdf['Year']==2016]
```

```python
mdf16.head()
```

```python
mdf16.tail()
```

```python
mdf16.Year.nunique()
```

- It is proved that the year column consists of only 2016 values hence proved

- We need to count the values for each sport so,

```python
mdf16m = pd.crosstab(index=mdf16['Sport'],columns='count')
```

```python
mdf16m
```

```python
mdf16m.sort_values(by='count',ascending=False)
```
